```python
import os
from openai import OpenAI
from pathlib import Pat
```

```python

## Aqui para is pruebas yo pase el API Key como un dato fijo
# y no como una variable de ambiente
#Este es el Comando del API_Key como variable de ambiente
# $env:OPENAI_API_KEY="sk-Aqui el API Key"
###api_key= "sk"
print(api_key)  
```

```python
client = OpenAI(api_key=api_key)
```

```python
def resumir_archivo(ruta_archivo: str) -> str:
    """
    Aqui subo cualuier tipo de Archivo a la API de OpenIA.
    """
    ruta = Path(ruta_archivo)
    if not ruta.is_file():
        raise FileNotFoundError(f"Archivo Inexistente: {ruta_archivo}")

    with open(ruta, "rb") as f:
        file_obj = client.files.create(
            file=f,
            purpose="user_data"   
        )

    response = client.responses.create(
        model="gpt-4.1", 
        input=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "input_text",
                        "text": (
                            "Resume este documento en español usando como máximo "
                            "Iindica tema principal, puntos claves "
                            "y conclusiones importantes."
                        ),
                    },
                    {
                        "type": "input_file",
                        "file_id": file_obj.id,  # aquí enlazamos el archivo subido :contentReference[oaicite:4]{index=4}
                    },
                ],
            }
        ],
        max_output_tokens=600,
    )
 
    resumen = response.output[0].content[0].text
    return resumen
```
