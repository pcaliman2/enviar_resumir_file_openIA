```python
from openai import OpenAI
from pathlib import Path

client = OpenAI()
```

```python
def resumir_archivo_texto(ruta_archivo: str, max_tokens_resumen: int = 400) -> str:
    """
    Lee un archivo de texto y devuelve un resumen generado por el modelo.
    """
    ruta = Path(ruta_archivo)
    if not ruta.is_file():
        raise FileNotFoundError(f"No se encontró el archivo: {ruta_archivo}")

    # Leer contenido
    texto = ruta.read_text(encoding="utf-8", errors="ignore")

    # Llamar al modelo para que lo resuma
    response = client.responses.create(
        model="gpt-5.1-mini",   
        input=(
            "Resume el siguiente texto en 10 viñetas como máximo, "
            "resaltando ideas principales y conclusiones:\n\n"
            f"{texto}"
        ),
        max_output_tokens=max_tokens_resumen,
    )

  
    resumen = response.output[0].content[0].text
    return resumen
```

```python
if __name__ == "__main__":
    resumen = resumir_archivo_texto("textotest.txt")
    print("=== RESUMEN ===")
    print(resumen)
```
